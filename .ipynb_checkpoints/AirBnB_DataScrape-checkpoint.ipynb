```python
import pandas as pd
import numpy as np

# Import Craigslist Data
craigs_df = pd.read_excel("Craigslist_Data.xlsx")

# Import AirBnB CSV
airbnb_df = pd.read_csv("AirBnB_SF.csv")
airbnb_df['Source'] = 'AirBnB'
airbnb_df.head()
```

```python
# Visualize Craigslist data and update AirBnB headers
craigs_df.head()
```

```python
airbnb_df = airbnb_df.rename(columns={'name': 'Post_Title', 'neighbourhood': 'Hood', 'price': 'Price'})
airbnb_df = airbnb_df[['Post_Title', 'Source', 'Hood', 'room_type', 'Price', 'latitude', 'longitude', 'minimum_nights',
                      'number_of_reviews']]
airbnb_df.head()
```

```python
# Join the 2 sets of data

housing_df = pd.concat([craigs_df, airbnb_df], ignore_index=True)
housing_df = housing_df[['Post_Title', 'Source', 'Hood', 'Price', 'Bedrooms', 'room_type', 'latitude', 'longitude', 'URL']]

#housing_df['Post_Title'].nunique()
housing_df.count()
```

```python
housing_df['Key'] = range(1, len(housing_df.index)+1)
housing_df['key'] = "a_"
housing_df['key'] = housing_df['key'] + housing_df['Key'].map(str)
housing_df = housing_df[['Post_Title', 'Source', 'Hood', 'Price', 'Bedrooms', 'room_type', 'latitude', 'longitude', 'URL', 'key']]
housing_df = housing_df.set_index('key')

housing_df
```

```python
housing_dict = dict(housing_df.to_dict())
print(housing_dict)
```
